```python
sales_dict={'apple':100,'banana':95,'oranges':85,'limes':99}
```

```python
bar_list=plt.bar(sales_dict.keys(),height=sales_dict.values(),color=['orange','red','green','blue'])
plt.axhline(25,color='black')
plt.xlabel('Fruits')
plt.grid()
```

```python
bar_list=plt.barh(list(sales_dict.keys()),list(sales_dict.values()))
plt.axvline(25,color='black')
plt.xlabel('Fruits')
plt.grid()
[plt.text(y,x,y) for x,y in sales_dict.items()]
plt.show()
```
