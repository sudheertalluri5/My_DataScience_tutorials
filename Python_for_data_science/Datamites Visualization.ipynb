```python
plt.plot(x,y)
```

```python
fig=plt.figure()
ax=fig.add_axes([0,0,0.5,0.5])
ax.plot(x,y)
# control limits of graph
ax.set_xlim([1,9])
ax.set_ylim([-1,10])

# display ticks
ax.set_xticks(np.arange(0,11))
ax.set_yticks(np.arange(0,10))
```

# Generating Mulitple plots

# Annotations

```python
mpg=mtcars.mpg
fig=plt.figure()
ax=fig.add_axes([0.1,1,2,1.2])
ax.set_xticks(range(len(mpg)))
ax.set_xticklabels(mtcars.car_name,rotation=90,fontsize=15)
ax.set_title('miles per galllon',fontsize=25)
ax.set_xlabel('car names',fontsize=15)
ax.set_xlabel('miles/gallon',fontsize=25)
ax.annotate('audi a4',xy=(19,31.9),xytext=(25,33),arrowprops=dict(facecolor='red',width=10,shrink=0.12))
ax.plot(mpg)
```
